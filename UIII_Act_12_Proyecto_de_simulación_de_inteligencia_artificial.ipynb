```python
# --- SISTEMA DE IA: GESTIÓN DE FOTÓGRAFOS 'PHOTO-GENIUS 2026' ---

# Definición de la clase Fotografo [3, 4]
class Fotografo:
    # El método __init__ es el constructor que inicializa los atributos del objeto [3, 5]
    def __init__(self, nombre, camara, lentes, especialidad, estudio, portafolio, resolucion):
        self.nombre = nombre
        self.tipo_camara = camara
        self.lentes_disponibles = lentes
        self.especialidad = especialidad # Atributo de especialidad (bodas/moda) [User Query]
        self.estudio = estudio
        self.portafolio = portafolio
        self.resolucion_salida = resolucion

    # Método de instancia que simula una IA analizando el equipo [6, 7]
    def analizar_perfil_ia(self):
        # La IA toma decisiones basadas en lógica condicional y relacional [8, 9]
        if self.resolucion_salida >= 45:
            resultado = "IA: Equipo de alta gama detectado. Apto para impresiones publicitarias."
        elif self.resolucion_salida >= 24:
            resultado = "IA: Resolución estándar profesional. Recomendado para eventos sociales."
        else:
            resultado = "IA: Resolución básica. Ideal para contenido web y redes sociales."

        # Sugerencia basada en especialidad utilizando operadores lógicos [8, 10]
        if self.especialidad == "bodas" and self.lentes_disponibles > 2:
            resultado += " Recomendación: Ideal para capturar ceremonias complejas."
        elif self.especialidad == "moda" or self.estudio == "si":
            resultado += " Recomendación: Perfil óptimo para sesiones editoriales en estudio."

        return resultado

# --- FUNCIONES DE GESTIÓN DEL SISTEMA ---

# Lista global que servirá como base de datos [11, 12]
galeria_fotografos = []

def agregar_fotografo():
    print("--- REGISTRO DE NUEVO FOTÓGRAFO ---")
    nombre = input("Nombre del fotógrafo: ")
    camara = input("Tipo de cámara (Reflex/Mirrorless): ")
    lentes = int(input("Cantidad de lentes disponibles: ")) # Entrada de datos interactiva [13, 14]
    especialidad = input("Especialidad (bodas/moda): ")
    estudio = input("¿Cuenta con estudio físico? (si/no): ")
    portafolio = input("Enlace o nombre del portafolio: ")
    resolucion = int(input("Resolución de salida (Megapíxeles): "))

    # Se crea la instancia y se añade a la lista [15, 16]
    nuevo = Fotografo(nombre, camara, lentes, especialidad, estudio, portafolio, resolucion)
    galeria_fotografos.append(nuevo)
    print(f"\nSISTEMA: {nombre} ha sido agregado exitosamente.")

def listar_fotografos():
    # Presentación en forma de tabla bien alineada [17, 18]
    if len(galeria_fotografos) == 0:
        print("SISTEMA: No hay fotógrafos registrados.")
        return

    print("\n" + "="*115)
    # Encabezados con formato de alineación (: < para izquierda, : > para derecha) [17]
    print(f"{'NOMBRE':<15} | {'CÁMARA':<12} | {'LENTES':<6} | {'ESPECIALIDAD':<12} | {'ESTUDIO':<7} | {'RESOLUCIÓN':<10} | {'ANÁLISIS IA'}")
    print("-" * 115)

    for f in galeria_fotografos: # Ciclo for para recorrer la secuencia de objetos [19, 20]
        ia_msg = f.analizar_perfil_ia()
        # Se accede a los atributos usando la notación de punto [21, 22]
        print(f"{f.nombre:<15} | {f.tipo_camara:<12} | {f.lentes_disponibles:<6} | {f.especialidad:<12} | {f.estudio:<7} | {f.resolucion_salida:<10} | {ia_msg[:40]}...")
    print("="*115 + "\n")

def actualizar_fotografo():
    nombre_buscado = input("Ingrese el nombre del fotógrafo a actualizar: ")
    encontrado = False
    for f in galeria_fotografos:
        if f.nombre == nombre_buscado:
            print(f"Modificando datos de {f.nombre}...")
            f.especialidad = input("Nueva especialidad: ")
            f.resolucion_salida = int(input("Nueva resolución de salida (MP): "))
            encontrado = True
            print("SISTEMA: Información actualizada.")
            break
    if not encontrado:
        print("SISTEMA: Fotógrafo no encontrado.")

def borrar_fotografo():
    nombre_buscado = input("Nombre del fotógrafo a eliminar: ")
    for i in range(len(galeria_fotografos)):
        if galeria_fotografos[i].nombre == nombre_buscado:
            # Se elimina el objeto de la lista [23, 24]
            del galeria_fotografos[i]
            print(f"SISTEMA: {nombre_buscado} ha sido borrado de la base de datos.")
            return
    print("SISTEMA: No se encontró al fotógrafo.")

# --- MENÚ PRINCIPAL INTERACTIVO ---

while True: # Bucle para mantener el menú activo [25, 26]
    print("\n--- MENÚ DE INTELIGENCIA FOTOGRÁFICA 2026 ---")
    print("1. Agregar Fotógrafo")
    print("2. Listar Fotógrafos (Ver Tabla)")
    print("3. Actualizar Datos")
    print("4. Borrar Fotógrafo")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_fotografo()
    elif opcion == "2":
        listar_fotografos()
    elif opcion == "3":
        actualizar_fotografo()
    elif opcion == "4":
        borrar_fotografo()
    elif opcion == "5":
        print("Cerrando sistema... ¡Buenas fotos!")
        break # Rompe el ciclo para finalizar el programa [26, 27]
    else:
        print("Opción inválida, intente de nuevo.")
```
